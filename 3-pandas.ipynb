### 2.3 数据输入输出

```python
# 模拟薪资数据
df = pd.DataFrame(np.random.randint(0, 100, size=[50, 5]),
                  columns=['IT', '化工', '生物', '教师', '士兵'])
df
```

```python
# 写入 csv文件
df.to_csv('./data.csv', sep=',', header=True, index=False)
```

```python
# 加载csv
pd.read_csv('./data.csv', sep=',', header=[0], index_col=None)
```

```python
# 写入 excel
df.to_excel('./data.xlsx', index=False, sheet_name='data')
```

```python
# 读取 excel
# pd.read_excel('./data.xlsx', sheet_name='data')
df2 = pd.read_excel('./data.xlsx', sheet_name=0, names=list('ABCDE'))
df2
```

```python
with pd.ExcelWriter('./data2.xlsx') as writer:
    df.to_excel(writer, index=False, sheet_name='data')
    df2.to_excel(writer, index=False, sheet_name='data2')
pd.read_excel('./data2.xlsx', sheet_name=1, )
```

### 2.4 数据选择

```python
df = pd.DataFrame(np.random.randint(0, 100, size=[1000, 3]), columns=['Python', 'Java', 'PHP'])
df
```

```python
df['Python']
df.Python
df[['Python', 'Java']]
df[['Python']]
```

```python

df = pd.DataFrame(np.random.randint(0, 100, size=[5, 3]),
                  index=list('ABCDE', ),
                  columns=['Python', 'Java', 'PHP'])
df
```

```python
# 读取行
df.loc['A']
df.loc[['A', 'B']]
```

```python
# 数字索引
df.iloc[[0, 1]]
```

```python
# 下面四种方法等效
df['Python']['B']
df.loc['B']['Python']
df.iloc[1]['Python']
df.iloc[1, 0]
```

```python
# 切片
df.loc['A':'C', 'Python':'Java']
```

```python

df.iloc[0:3, 0:2]
```

#### Boolean 索引

```python
cond = df.Python > 50
cond
```

```python
df[cond]
```

```python
# 筛选Python > 50 并且 PHP > 50 的行
cond = (df.Python > 50) & (df.PHP > 50)
df[cond]
```
